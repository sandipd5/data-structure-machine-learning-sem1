### Q5. Drop duplicate rows, if any. If so, keep the first duplicate of each such duplicate set

```python
duplicated_rows = fire_df.duplicated()   #check duplicated rows
print(fire_df[duplicated_rows])
duplicated_rows = fire_df.duplicated().sum()  #count no of duplicated rows
duplicated_rows
```

```python
# Drop duplicate rows while keeping the first occurrence
fire_df.drop_duplicates(keep = 'first', inplace = True)
# Reset the index of the DataFrame
fire_df.reset_index(drop=True, inplace=True)
fire_df
```

```python
duplicated_rows = fire_df.duplicated()   # Again check duplicated rows
print(fire_df[duplicated_rows])
duplicated_rows = fire_df.duplicated().sum()  #Again count no of duplicated rows
duplicated_rows    #from output we see there is no duplicates rows now
```

### (II) Prepare the data:
 ### Add the `fire_month` and `days_burning` columns


```python
fire_miss_df['contain_date'].info()
```

```python
fire_miss_df['fire_month'] = fire_miss_df['contain_date'].dt.month

fire_miss_df
```

```python

fire_miss_df['days_burning'] = (fire_miss_df['contain_date'] - fire_miss_df['discovery_date']).dt.days
fire_miss_df
```
