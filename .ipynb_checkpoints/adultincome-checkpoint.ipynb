### We have some missing values in form of ? in workclass, occupation and country column.
>We will handle them separately either by imputing or removing those rows

```python
# Calculate value counts for workclass
value_counts = income_df['workclass'].value_counts()

# Normalize value counts to percentages
normalized_value_counts = (value_counts / value_counts.sum()) * 100
normalized_value_counts
```

```python
# Calculate value counts for occupation
value_counts = income_df['occupation'].value_counts()

# Normalize value counts to percentages
normalized_value_counts = (value_counts / value_counts.sum()) * 100
normalized_value_counts
```

```python
# Calculate value counts for country
value_counts = income_df['native.country'].value_counts()

# Normalize value counts to percentages
normalized_value_counts = (value_counts / value_counts.sum()) * 100
normalized_value_counts
```

### Custom mapping income variable

```python
income_mapping = {'<=50K': 0, '>50K': 1}
income_df['income'] = income_df['income'].map(income_mapping)
```

```python
income_df
```

### Finding relations between income and numerical features
> We are using point-biserial correlation, which is appropriate for measuring the strength and direction of the association between a binary (0 or 1) variable and a continuous variable. In this case, we can calculate the point-biserial correlation coefficient using the scipy library.

```python
from scipy.stats import pointbiserialr

# Calculate point-biserial correlation
correlation_matrix = {}
#for column in income_df[categorical_columns]:
correlation, p_value = pointbiserialr(income_df['income'], income_df[categorical_columns])
correlation_matrix[categorical_columns] = correlation
print("Point-Biserial Correlation Matrix:\n", correlation_matrix)
```
