## Basics of Dictionary

```python
## Basics of Dictionary
dict_1 = {"ontario":"Toronto","Manitoba":"winipeg",
         "BC":"vancouver"}
dict_1
```

```python
for i in dict_1.items():
    print(i)
```

```python
for k,v in dict_1.items():
    print(v)
```

```python
dict_2 = {"ontario":["Toronto","Abbotsford"],"Manitoba":["Winipeg","Brandon"],"BC":["Vancouver","Victoria","Vernon"]}
dict_2.items()
```

```python
for k,v in dict_2.items():
    print(v[1])
```
