##### 6.1) How to drop missing values:

```python
mort_df.dropna()
```

```python
mort_df.shape[0] - mort_df[mort_df.isnull().any(axis = 1)].shape[0]

# 476-5
```

```python
# Rows which contain 2 or more missing values:
mort_df[mort_df.isnull().sum(axis = 1) >= 2]
```

```python
# Drop the rows that have 2 or more missing values:
mort_df.dropna(thresh = 2, inplace = False)
```

```python
# Drop the rows that have missing values, in the ['DeathRate'] column:
mort_df.dropna(subset = ['DeathRate'], inplace = False)  
```

##### 6.2) How to fill missing values:

```python
mort_df.head()
```

```python
# Mean imputation:
mort_df['DeathRate'].fillna(value = mort_df['DeathRate'].mean(), inplace = True)
```

```python
mort_df.head()
```

```python
mort_df.isnull().sum()
```

```python
# Forward-fill:
mort_df.fillna(method = "ffill", limit = 2, inplace = True)
```

```python
mort_df.isnull().sum()
```

#### 8.2) Handling outliers:

```python
teen_df['DeathRate'].mean()
```

```python
teen_df.head()
```

```python
# rows with excess (ie, outlier) DeathRate:
teen_df.loc[teen_df['DeathRate'] > 450] ['DeathRate'] 

#= teen_df['DeathRate'].mean()
```

```python
teen_df.loc[teen_df['DeathRate'] > 500] ['DeathRate'] 
```

```python
# 1) replacing the value of the outlier, with the threshold:
teen_df.loc[teen_df['DeathRate'] > 500, 'DeathRate'] = 500
```

```python
# checking whether the outlier '777' has indeed been replaced:
teen_df[teen_df['DeathRate'] >= 500]
```

```python
#2) replacing the outlier by the Mean:
teen_df.loc[teen_df['DeathRate'] == 500, 'DeathRate'] = teen_df['DeathRate'].mean()
```

```python
# checking that the outlier 'DeathRate' has indeed by replaced by the Mean of the DeathRate:
teen_df[teen_df['Year'] == 1918]
```
