## Finding outliers

```python
teen_df = mor_df[mor_df['AgeGroup'] == '15-19 Years']      #select teen age group and put it in new variable
teen_df.shape
```

```python
teen_df.describe()     #describe will give all statistics parameters
```

```python
teen_df.describe().T       #max is 777.4 which is far from 75% quantiles , so we will see if outlies exists
```

```python
teen_df[teen_df['DeathRate'] > 500]  # we see there is only 1 value greater than 500 and that is 777.4 itself, 
#500 is taken as threshold between value of 283.65 and 777.4
```

```python
#using plottting to find outliers
teen_df.plot(x='Year',y ='DeathRate')  #we see there is surge in death i.e 777.4 in 1918
```

## Handling Outliers

```python
#replacing the outlier with outlier that is 777 with 500
teen_df.loc[teen_df['DeathRate'] > 500, 'DeathRate'] = 500  
```

```python
#checking if 777 is replaced
teen_df[teen_df['DeathRate'] >= 500]
```

```python
#replacing outlier by mean
teen_df.loc[teen_df['DeathRate'] == 500, 'DeathRate'] = teen_df['DeathRate'].mean()
```
